```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
url = "https://microdata.worldbank.org/index.php/api/tables/info/fcv/wld_2021_rtfp_v02_m"
```

```python
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    try:
        # Parse the JSON response
        data = response.json()

        # Print the data in a readable format
        print("JSON Data Retrieved Successfully:")
        print(json.dumps(data, indent=4))

        # Save the JSON data to a file
        with open("output.json", "w") as file:
            json.dump(data, file, indent=4)
        
        # Convert JSON to DataFrame and save as CSV
        if isinstance(data, list):  # If data is a list of dictionaries
            df = pd.DataFrame(data)
        elif isinstance(data, dict):  # If data is a dictionary
            df = pd.DataFrame.from_dict(data, orient="index").reset_index()
        else:
            print("Unsupported JSON structure for CSV conversion.")
            df = None

        if df is not None:
            # Save the DataFrame to a CSV file
            csv_file = "output.csv"
            df.to_csv(csv_file, index=False)
            print(f"Data saved to {csv_file}")

    except ValueError:
        print("Failed to parse JSON. The response might not be valid JSON.")
else:
    print(f"Failed to fetch data. HTTP Status Code: {response.status_code}")
```
